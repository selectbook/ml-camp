### 2.PySparkå®éªŒé¢˜(45åˆ†)
åœ¨æœåŠ¡å™¨ä¸Šæ‰§è¡Œpysparkå‘½ä»¤å¯ä»¥å¯åŠ¨ï¼Œè¯·æŠŠdata.csvæ–‡ä»¶æ”¾ç½®åœ¨æœåŠ¡å™¨ä¸Š

##### 1.è½½å…¥æ•°æ®ä¸äº†è§£åŸºæœ¬ä¿¡æ¯

````python
>>> pyspark

from pyspark import SparkContext
sc = SparkContext('local', 'pyspark')
````

````python
# è¯»å…¥æ•°æ®åˆ°raw_contentè¿™ä¸ªRDDä¹‹ä¸­
raw_content = sc.textFile('data.csv')
````

````python
# é€šè¿‡å‘½ä»¤ç»Ÿè®¡raw_contentä¸­çš„è®°å½•æ¡æ•°
raw_content.count()
````

````python
# ä»raw_contentè¿™ä¸ªRDDä¸­å–å‡ºå‰5æ¡è®°å½•è¾“å‡º
raw_content.top(5)
````

````python
# ä»raw_contentä¸­é‡‡æ ·å‡º3æ¡è®°å½•è¾“å‡º
raw_content.takeSample(withReplacement=False, num=3)
````

##### 2.æ•°æ®å¤„ç†

````python
# å¯¹ä»¥ä¸Šæ•°æ®è¿›è¡Œæ¸…æ´—ï¼ŒæŠŠè®°å½•ä¸­å¸¦çš„å¼•å·å»é™¤æ‰ï¼Œå¾—åˆ°content_rdd
# tipï¼šå‚è€ƒmapå‡½æ•°çš„ç”¨æ³•
content_rdd = raw_content.map(lambda x : x.replace('"',''))
````

ä½ å°†å¾—åˆ°ç±»ä¼¼ä»¥ä¸‹ç»“æœçš„RDDï¼š
```
[
[u'date', u'time', u'size', u'r_version', u'r_arch', u'r_os', u'package', u'version', u'country', u'ip_id'], 
[u'2015-12-12', u'13:42:10', u'257886', u'3.2.2', u'i386', u'mingw32', u'HistData', u'0.7-6', u'CZ', u'1'], 
[u'2015-12-12', u'13:24:37', u'1236751', u'3.2.2', u'x86_64', u'mingw32', u'RJSONIO', u'1.3-0', u'DE', u'2'], 
[u'2015-12-12', u'13:42:35', u'2077876', u'3.2.2', u'i386', u'mingw32', u'UsingR', u'2.0-5', u'CZ', u'1']
]
```

````python
# æœ‰å¦‚ä¸‹çš„textï¼Œè¯·ä½¿ç”¨flatmapå¾—åˆ°åŒ…å«æ‰€æœ‰å­—å¹•çš„list
# tipï¼šè¯·å‚è€ƒè¯¾ç¨‹flatmapçš„ä½¿ç”¨ï¼Œç‰¹åˆ«æ³¨æ„è¿™é‡Œä¸åŒåˆ†å‰²ç¬¦çš„å¤„ç†
import re

text=["a b c", "d e", "f#g#h", "m n q", "r,q,w", "j%r%q"]

rdd = sc.parallelize(text)
print(rdd.flatMap(lambda s:re.split(' |\#|,|%',s)).collect())
````

ä½ å°†å¾—åˆ°å¦‚ä¸‹ç»“æœçš„RDDï¼š
```
[a, b, c, d, e, f, g, h, m, n, q, r, q, w, j, r, q]
```

````python
# ä»åˆšæ‰çš„content_rddä¸­å–å‡ºç¬¬7åˆ—ï¼Œå¯¹ä¸åŒçš„packageç±»åˆ«è¿›è¡Œç»Ÿè®¡è®¡æ•°
# tipï¼šå¯ä»¥ä½¿ç”¨map reduceæˆ–è€…pair-rdd reduceByKey
content_rdd = content_rdd.map(lambda x: x.split(","))                        
content_rdd.map(lambda x: (x[6], 1)).reduceByKey(lambda x, y: x+y).collect() 
````

##### 3.æ’åº

````python
# å–å‡ºæ•°é‡æœ€å¤šçš„10ä¸ªpackageå’Œå‡ºç°çš„é¢‘æ¬¡
# tipï¼šæ³¨æ„sortbykeyçš„ä½¿ç”¨
content_rdd.map(lambda x: (x[6], 1)).reduceByKey(lambda x, y: x+y).sortBy(lambda x:x[1],False).take(10)
````

##### 4.è¿‡æ»¤

````python
# å–å‡ºtop5çš„packageå¯¹åº”çš„æ•°æ®è®°å½•
# tipï¼šæ³¨æ„filterçš„ä½¿ç”¨
# é¦–å…ˆå–å‰äº”çš„package
top5 = content_rdd.map(lambda x: (x[6], 1)).reduceByKey(lambda x, y: x+y).sortBy(lambda x:x[1],False).take(5)  
# å°† top5 ä¸­packageåç»„åˆæˆlistï¼štop5_pkg
top5_pkg = list(map(lambda x:x[0],top5))   
# æ ¹æ®top5_pkgè¿‡æ»¤å…¨éƒ¨æ•°æ®é›†
top5_items = content_rdd.filter(lambda x:True if x[6] in top5_pkg else False).collect()   
# top5çš„æ•°é‡ï¼š19329
print(len(top5_items)) 
# è¾“å‡ºå«æœ‰top5çš„å…¨éƒ¨æ•°æ®
print(top5_items) 

````
